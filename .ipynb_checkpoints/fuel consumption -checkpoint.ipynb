```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("fuel consumption.csv")
```

```python
df.info()
```

Model

    4WD/4X4 = Four-wheel drive
    AWD = All-wheel drive
    CNG = Compressed natural gas
    FFV = Flexible-fuel vehicle
    NGV = Natural gas vehicle
    # = High output engine that provides more power than the standard engine of the same size

Transmission

    A = Automatic
    AM = Automated manual
    AS = Automatic with select shift
    AV = Continuously variable
    M = Manual
    3 - 10 = Number of gears
    Fuel Type
    X = Regular gasoline
    Z = Premium gasoline
    D = Diesel
    E = Ethanol (E85)
    N = Natural Gas

Fuel Consumption
City and highway fuel consumption ratings are shown in litres per 100 kilometres (L/100 km) - combined rating (55% city, 45% hwy) is shown in L/100 km and in miles per imperial gallon (mpg)
CO2 Emissions (g/km),Estimated tailpipe carbon dioxide emissions (in grams per kilometre) are based on fuel type and the combined fuel consumption rating.

# UNIVARIATE ANALYSIS

## Histogram

***Histogram contains a group of bars that show the density of the data***

```python
sns.histplot(x="COMB (mpg)",data=df,color="red",bins=7,kde=True)
plt.xlabel("Combustion");
plt.ylabel("Count")

# kde = The KDE plot is the method used to visualize the distribution of the dataset.
```

## Bar Chart (Count Plot)

***A bar chart is a way of summarizing a set of categorical data***

```python
df["FUEL"].unique()
```

```python
sns.countplot(x="FUEL",data=df);
```

# BIVARIATE ANALYSIS

## Barplot (Categorical v/s Numerical)

```python
plt.figure(figsize=(7,5))
sns.barplot(x="FUEL",y="FUEL CONSUMPTION",data=df);
plt.xlabel("Fuel")
plt.ylabel("Fuel Consumption")
plt.title("Fuel Consumption w.r.t fuel types")
```

## Violinplot

```python
sns.violinplot(x="FUEL",y="FUEL CONSUMPTION",data=df)
```

## Scatterplot (Numerical v/s Numerical)

```python
sns.scatterplot(x="CYLINDERS",y="EMISSIONS",data=df,alpha=0.7);
```

```python
sns.scatterplot(x="HWY (L/100 km)",y="FUEL CONSUMPTION",data=df)
```
